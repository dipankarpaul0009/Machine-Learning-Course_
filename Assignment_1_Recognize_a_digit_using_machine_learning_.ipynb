```python
#### Recognize a Digit using Machine Learning ####

# ----------------------------------------------------------
# 1. Import Required Libraries
# ----------------------------------------------------------
# TensorFlow: for building and training the neural network
import tensorflow as tf

# Layers and model-building tools from Keras API
from tensorflow.keras import layers, models

# MNIST dataset from Keras (handwritten digits 0–9)
from tensorflow.keras.datasets import mnist

# Matplotlib: to display images
import matplotlib.pyplot as plt

# Numpy: for numerical operations
import numpy as np


# ----------------------------------------------------------
# 2. Load the MNIST Dataset
# ----------------------------------------------------------
# The dataset is split into:
# - Training data: x_train (images), y_train (labels)
# - Testing data: x_test (images), y_test (labels)
(x_train, y_train), (x_test, y_test) = mnist.load_data()


# ----------------------------------------------------------
# 3. Normalize the Dataset
# ----------------------------------------------------------
# Pixel values in MNIST range from 0 to 255 (grayscale).
# Dividing by 255.0 scales values to 0–1, which speeds up training.
x_train = x_train / 255.0
x_test = x_test / 255.0


# ----------------------------------------------------------
# 4. Build the Neural Network Model
# ----------------------------------------------------------
# We'll use a simple feedforward neural network:
# Flatten → Dense(128, relu) → Dense(10, softmax)
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),      # Convert 28×28 images to a 1D vector
    layers.Dense(128, activation='relu'),      # Hidden layer with 128 neurons
    layers.Dense(10, activation='softmax')     # Output layer with 10 neurons (digits 0–9)
])


# ----------------------------------------------------------
# 5. Compile the Model
# ----------------------------------------------------------
# - optimizer='adam': Adaptive learning rate optimization
# - loss='sparse_categorical_crossentropy': for multi-class classification
# - metrics=['accuracy']: track accuracy during training/testing
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])


# ----------------------------------------------------------
# 6. Train the Model
# ----------------------------------------------------------
# epochs=5 → The training will go through the dataset 5 times
model.fit(x_train, y_train, epochs=5)


# ----------------------------------------------------------
# 7. Evaluate the Model
# ----------------------------------------------------------
# Check performance on the unseen test dataset
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f"Test Accuracy: {test_acc:.4f}")


# ----------------------------------------------------------
# 8. Pick a Test Image of Digit 9
# ----------------------------------------------------------
# Find the index of the first image labeled as '9' in y_test
index_of_nine = np.where(y_test == 9)[0][0]

# Display the image
plt.imshow(x_test[index_of_nine], cmap='gray')
plt.title(f"Actual: {y_test[index_of_nine]}")  # Should display "Actual: 9"
plt.show()


# ----------------------------------------------------------
# 9. Predict the Digit
# ----------------------------------------------------------
# Model expects a batch (array of images), so wrap the single image in an array
prediction = model.predict(np.array([x_test[index_of_nine]]))

# The output is a probability distribution → Pick the index with the highest probability
predicted_label = tf.argmax(prediction[0]).numpy()

# Display the predicted label
print(f"Predicted Label: {predicted_label}")
```
